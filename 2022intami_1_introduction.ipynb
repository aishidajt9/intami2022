## random walk

````python
import matplotlib.pyplot as plt
import random
````

1本のランダム・ウォークを描画する．

````python
x = 0
ls = []
ls.append(x)

for i in range(10000):
    x += random.randint(-1, 1)
    ls.append(x)

plt.plot(ls)
plt.show();
````

100本のランダム・ウォークを描画する．

````python
lss = []
for j in range(100):
    x = 0
    ls = []
    ls.append(x)

    for i in range(1000):
        x += random.randint(-1, 1)
        ls.append(x)
    
    lss.append(ls)

for ls in lss:
    plt.plot(ls)
plt.show();
````

最後の位置のヒストグラム

````python
lastlocation = [ls[-1] for ls in lss]
plt.hist(lastlocation)
plt.show();
````
